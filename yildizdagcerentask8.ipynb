```python
# JokeTexts data
dfJoke = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/JokeText.csv')

# User ratings for each joke
dfRtg1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/UserRatings1.csv')
dfRtg2 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/UserRatings2.csv')
```

```python
# Remove duplicates
dfJoke.drop_duplicates(subset ='JokeText', keep = 'first', inplace = True)
dfJoke.shape
```

```python
# Generate a matrix of common terms that show up in each joke

from sklearn.feature_extraction.text import TfidfVectorizer
mdlTfvMvs = TfidfVectorizer(analyzer='word',ngram_range=(1, 2), stop_words='english')
tfidf_matrix = mdlTfvMvs.fit_transform(dfJoke['JokeText'])
tfidf_matrix.shape
```

```python
# Calculate cosine similarity between each pair of jokes as a function of the similarity of the common terms

from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
cosine_sim.shape
```

```python
# Prepare recommendation function
def get_recommendations(JokeId):
    sim_scores = list(enumerate(cosine_sim[JokeId]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:31]
    joke_indices = [i[0] for i in sim_scores]
    return dfJoke.iloc[joke_indices]
```

```python
dfRatings = pd.concat([dfRtg1, dfRtg2], ignore_index=True)
```

```python
from surprise import Reader, Dataset
from surprise.model_selection import train_test_split
```

```python
# Reshape the data into long format
df_long = dfRatings.melt(id_vars=['JokeId'], var_name='User', value_name='Rating')

# Filter out rows with missing ratings (if any)
df_long = df_long.dropna(subset=['Rating'])

# Convert JokeId and User columns to strings (if necessary)
df_long['JokeId'] = df_long['JokeId'].astype(str)
df_long['User'] = df_long['User'].astype(str)

reader = Reader(rating_scale=(0, 5))

X = Dataset.load_from_df(df_long[['User', 'JokeId', 'Rating']], reader)
X_train, X_test = train_test_split(X, test_size=.25)

print(df_long)
```

```python
# Define SVD model

from surprise import SVD, accuracy

mdlSvdRtg = SVD()
```

```python
# Fit SVD model

mdlSvdRtg.fit(X_train)
test_pred = mdlSvdRtg.test(X_test)
```

```python
# Evalute SVD accuracy

accuracy.rmse(test_pred)
```

```python
# Tune hyperparameters

from surprise.model_selection import GridSearchCV

param_grid = {'n_epochs': [5, 10, 15], 'lr_all': [0.002, 0.005],
              'reg_all': [0.4, 0.6]}
gs = GridSearchCV(SVD, param_grid, measures=['rmse', 'mae'], cv=3)

gs.fit(X)

# best RMSE score
print(gs.best_score['rmse'])

# combination of parameters that gave the best RMSE score
print(gs.best_params['rmse'])
```

```python
# Cross-validate

from surprise.model_selection import cross_validate

cross_validate(mdlSvdRtg, X, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```
