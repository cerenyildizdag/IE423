Moreover, let's draw the graph of the average yellow and red card penalties according to minutes played.

```python
import pandas as pd

# Split the 'minutes played' column into 100 minute intervals
dfFootball['minutes_group'] = pd.cut(dfFootball['minutes played'], bins=range(0, dfFootball['minutes played'].max() + 100, 100))

# Calculate the average of card numbers for each group
cards_by_minutes = dfFootball.groupby('minutes_group').agg({'yellow cards': 'mean', 'second yellow cards': 'mean', 'red cards': 'mean'})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.lineplot(x=cards_by_minutes.index.astype(str), y='yellow cards', data=cards_by_minutes, label='Yellow Cards')
sns.lineplot(x=cards_by_minutes.index.astype(str), y='second yellow cards', data=cards_by_minutes, label='Second Yellow Cards')
sns.lineplot(x=cards_by_minutes.index.astype(str), y='red cards', data=cards_by_minutes, label='Red Cards')

plt.xlabel('Minutes Played (Groups)')
plt.ylabel('Average Card Count')
plt.title('Cards by Minutes Played')
plt.legend()
plt.show()
```

**Key Observations**:

* **Defender Centre-Back**: There is a wide range of market values, with several players valued highly.
* **Attack Centre-Forward**: This position also shows a wide range of values, with some players having extremely high market values.
* **Goalkeeper**: Goalkeepers show a narrower distribution of market values compared to outfield players.
* **Midfield Positions**: Midfielders have a significant spread in market values, with "midfield-CentralMidfield" and "midfield-AttackingMidfield" having a notable number of highly valued players.
* **Attack Positions**: Similar to midfielders, players in attacking positions like "Attack-RightWinger" and "Attack-LeftWinger" show a wide range of market values.
* **Yellow Cards**: Players tend to receive more yellow
cards at the beginning of the match, with the count decreasing as the game progresses.

* **Second Yellow Cards** **and Red Cards**: Both types of cards are given much less frequently compared to yellow cards. Their counts remain relatively constant across different minutes played.
